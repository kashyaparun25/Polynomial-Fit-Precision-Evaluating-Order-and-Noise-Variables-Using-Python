### Q1. Please plot the noisy data and the polynomial you found (in the same figure). Please use polynomial order of m = 1, 2, 3, 4, 5, 6, 7, 8 respectively.

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

# Generates noisy data (given in the question)
noise_scale = 100
number_of_samples = 50
x = 30 * (np.random.rand(number_of_samples, 1) - 0.5)
y = 2 * x + 11 * x**2 + 3 * x**3 + noise_scale * np.random.randn(number_of_samples, 1)

# Creates an array of x values for plotting the polynomial fits
x_fit = np.linspace(min(x), max(x), 100)

# Creates a figure to display the subplots and change the sizes of the subplots
plt.figure(figsize=(15, 7))

# Iterates through polynomial orders from 1 to 8
for m in range(1, 9):
    # Fit a polynomial of order 'm' to the data
    coefficients = np.polyfit(x.flatten(), y.flatten(), m)

    # Evaluates the polynomial at the x_fit values
    y_fit = np.polyval(coefficients, x_fit)

    # Creates a subplot for each polynomial order with given rows and columns
    plt.subplot(2, 4, m)

    # Plots the noisy data points on the graph and labeling the plot pointer used
    plt.plot(x, y, 'ro', label='Noisy Data')

    # Plots the polynomial fit
    plt.plot(x_fit, y_fit, label=f'Order {m} Polynomial Fit')

    # Adds a legend to the plot which is required to display legends
    plt.legend()

plt.tight_layout()
plt.show()
```

### Q2. Plot MSE versus order m, for m = 1, 2, 3, 4, 5, 6, 7, 8 respectively. Identify the best choice of m

```python
# Import necessary libraries
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
import numpy as np  # Import NumPy for numerical operations

# Generates noisy data
noise_scale = 100  # Set the scale of the noise in the data
number_of_samples = 50  # Determine the number of data points
x = 30 * (np.random.rand(number_of_samples, 1) - 0.5)  # Generate random x values
y = 2 * x + 11 * x**2 + 3 * x**3 + noise_scale * np.random.randn(number_of_samples, 1)  # Generate noisy y values

# Creates an array of x values for plotting the polynomial fits
x_fit = np.linspace(min(x), max(x), 100)  # Create a range of x values for smooth plotting

# Initialize an empty list to store MSE values
mse_values = []

# Iterates through different polynomial orders from 1 to 8
for m in range(1, 9):
    # Fit a polynomial of order 'm' to the data
    coefficients = np.polyfit(x.flatten(), y.flatten(), m)  # Perform polynomial fitting

    # Evaluate the polynomial at the x values
    y_fit = np.polyval(coefficients, x)  # Calculate fitted y values

    # Calculate the Mean Squared Error (MSE)
    mse = np.mean((y - y_fit)**2)  # Compute the MSE

    # Append the MSE to the list for later analysis
    mse_values.append(mse)  # Store the MSE value in a list
    print(f"For the Polynomial of Order {m}, The MSE is {mse:.2f}")  # Print the MSE for each order

# Plot MSE versus polynomial order (m)
plt.figure(figsize=(8, 5))  # Create a plot with specified dimensions
plt.plot(range(1, 9), mse_values, marker='o', linestyle='-')  # Plot the MSE values
plt.xlabel('Polynomial Order (m)')  # Label for the x-axis
plt.ylabel('Mean Squared Error (MSE)')  # Label for the y-axis
plt.title('MSE versus Polynomial Order (m)')  # Title for the plot
plt.grid(True)  # Add gridlines to the plot

# Find the best choice of m (minimum MSE)
best_m = np.argmin(mse_values) + 1  # Identify the order with the lowest MSE (add 1 due to 0-based indexing)
min_mse = min(mse_values)  # Get the minimum MSE value

# Annotate the plot with information about the best choice of m and its corresponding MSE
plt.annotate(f'Best m={best_m}\nMSE={min_mse:.2f}', xy=(best_m, min_mse), xytext=(best_m + 1, min_mse * 2),
             arrowprops=dict(arrowstyle='->'))

plt.show()  # Display the plot

# Print the final result indicating the best choice of m and its corresponding MSE
print(f"The best choice of m is {best_m} with an MSE of {min_mse:.2f}")
```

### Q3. Change variable noise_scale to 200, 300, 400, 600, 800, 1000 respectively, re-run the algorithm and plot the polynomials with the best m found in 2). Discuss the impact of noise scale to the accuracy of the returned parameters. [You need to plot a figure for EACH choice of noise_scale.]

```python
# Import necessary libraries
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
import numpy as np  # Import NumPy for numerical operations
from numpy.polynomial.polynomial import polyfit  # Import a specific function for polynomial fitting

# Define different noise_scale values to investigate
noise_scales = [200, 300, 400, 600, 800, 1000]

# Create a figure for each noise_scale value
for noise_scale in noise_scales:
    # Generate noisy data
    number_of_samples = 50  # Define the number of data points
    x = 30 * (np.random.rand(number_of_samples, 1) - 0.5)  # Generate random x values
    y = 2 * x + 11 * x**2 + 3 * x**3 + noise_scale * np.random.randn(number_of_samples, 1)  # Generate noisy y values

    # Create an array of x values for smooth plotting of polynomial fits
    x_fit = np.linspace(min(x), max(x), 100)

    # Initialize an empty list to store MSE (Mean Squared Error) values
    mse_values = []

    # Iterate through different polynomial orders from 1 to 8
    for m in range(1,9):
        # Fit a polynomial of order 'm' to the noisy data
        coefficients = polyfit(x.flatten(), y.flatten(), m)

        # Evaluate the polynomial at the x values
        y_fit = np.polyval(coefficients, x)

        # Calculate the Mean Squared Error (MSE) between noisy data and fitted data
        mse = np.mean((y - y_fit)**2)

        # Append the computed MSE value to the list for later analysis
        mse_values.append(mse)

    # Find the best choice of polynomial order 'm' (minimum MSE)
    best_m = np.argmin(mse_values) + 1  # +1 to adjust for 0-based indexing
    best_mse = min(mse_values)  # Get the minimum MSE value

    # Create a new figure (plot) for the current noise_scale value
    plt.figure(figsize=(7, 5))

    # Plot the noisy data points in red
    plt.scatter(x, y, color='red', label='Noisy Data')

    # Fit the best polynomial to the noisy data
    best_coefficients = polyfit(x.flatten(), y.flatten(), best_m)
    y_best_fit = np.polyval(best_coefficients, x_fit)

    # Plot the best polynomial fit in blue
    plt.plot(x_fit, y_best_fit, color='blue', label=f'Best Order {best_m} Polynomial Fit')

    # Add labels and title to the plot
    plt.title(f'Noise Scale = {noise_scale}')
    plt.xlabel('x')
    plt.ylabel('y')

    # Add a legend to the plot
    plt.legend()

    # Print the best 'm' (polynomial order) and its corresponding MSE
    print(f'\nFor noise_scale = {noise_scale}, Best m = {best_m}, MSE = {best_mse:.2f}')

    # Display the plot for the current noise_scale value
    plt.show()
```

As noise_scale increases, the data becomes noisier, leading to greater data point variability in the plots. Higher noise_scale values make accurate polynomial fitting more challenging. The best m choice may differ for various noise scales, but typically, lower-order polynomials work better for higher noise_scales because they are less prone to overfitting. Balancing model complexity (polynomial order) and data noise is crucial for achieving accurate and generalizable results.

### Q4. Change variable number_of_samples to 40, 30, 20, 10 respectively, re-ran the algorithm and plot the polynomials with the best m found in 2). Discuss the impact of the number of samples to the accuracy of the returned parameters. [You need to plot a figure for EACH choice of number_of_samples.]

```python
# Import necessary libraries
import matplotlib.pyplot as plt  # Import Matplotlib for creating plots
import numpy as np  # Import NumPy for mathematical operations

# Define different sample_sizes to study
sample_sizes = [40, 30, 20, 10]

# Create a figure for each sample size
for num_samples in sample_sizes:
    # Create empty lists to store results for each sample size
    mse_values = []  # To store Mean Squared Error (MSE) values
    best_m_values = []  # To store the best polynomial order values

    # Define different noise_scale values
    noise_scales = [200, 300, 400, 600, 800, 1000]

    # Iterate through different noise_scale values
    for noise_scale in noise_scales:
        # Generate noisy data for the current sample size and noise scale
        x = 30 * (np.random.rand(num_samples, 1) - 0.5)  # Generate random x values
        y = 2 * x + 11 * x**2 + 3 * x**3 + noise_scale * np.random.randn(num_samples, 1)  # Generate noisy y values

        # Create an array of x values for smoother plotting of polynomial fits
        x_fit = np.linspace(min(x), max(x), 100)

        # Initialize an empty list to store MSE values specific to the current noise_scale
        mse_values_noise = []

        # Iterate through polynomial orders from 1 to 8
        for m in range(1,9):
            # Fit a polynomial of order 'm' to the noisy data
            coefficients = polyfit(x.flatten(), y.flatten(), m)

            # Evaluate the polynomial at the x values
            y_fit = np.polyval(coefficients, x)

            # Calculate the Mean Squared Error (MSE) between noisy data and fitted data
            mse = np.mean((y - y_fit)**2)

            # Append the computed MSE value to the list for later analysis
            mse_values_noise.append(mse)

        # Find the best choice of polynomial order 'm' (minimum MSE) for the current noise_scale
        best_m = np.argmin(mse_values_noise) + 1  # Adding 1 to account for 0-based indexing

        # Append the best 'm' and its corresponding MSE to the lists
        best_m_values.append(best_m)
        mse_values.append(mse_values_noise[best_m - 1])  # Subtracting 1 to get the MSE for the best 'm'

    # Create a figure to display the results for the current sample size
    plt.figure(figsize=(12, 5))

    # Plot MSE versus noise_scale
    plt.subplot(1, 2, 1)
    plt.plot(noise_scales, mse_values, marker='o', linestyle='-')
    plt.xlabel('Noise Scale')
    plt.ylabel('Mean Squared Error (MSE)')
    plt.title(f'MSE versus Noise Scale (Sample Size = {num_samples})')
    plt.grid(True)

    # Plot best 'm' versus noise_scale
    plt.subplot(1, 2, 2)
    plt.plot(noise_scales, best_m_values, marker='o', linestyle='-')
    plt.xlabel('Noise Scale')
    plt.ylabel('Best Polynomial Order (m)')
    plt.title(f'Best Polynomial Order versus Noise Scale (Sample Size = {num_samples})')
    plt.grid(True)

    plt.tight_layout()  # Ensures proper spacing between subplots

    # Display the plots for the current sample size
    plt.show()
```

Reducing the number of data samples means there's less information to create accurate polynomial fits. Smaller sample sizes can lead to greater fluctuations in the calculated polynomial characteristics. When there are fewer data points, it becomes harder for models to capture the core patterns in the data, especially when there's a lot of noise. The plotted data illustrates how noise_scale, mean squared error (MSE), and the optimal polynomial order change with different sample sizes. These visualizations show the impact of sample size on the precision of polynomial fitting, particularly when noise levels are high.
